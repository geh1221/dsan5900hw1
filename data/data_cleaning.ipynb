```python
import pandas as pd
import numpy as np

data = pd.read_csv("SummerStudentAdmissions2.csv")
data.head(100)
```

```python
df_filled = data.fillna("N/A")
df_filled
```

### Dropping all NA

```python
data = data.dropna()
data
```

### Decision Cleaning

```python
print("Decision Unique Values:", data['Decision'].unique())
```

```python
data = data[data["Decision"] != "Banana"]
```

```python
print("Decision Unique Values:", data['Decision'].unique())
```

### State

```python
print("State Unique Values:", data['State'].unique())
data["State"] = data["State"].str.title()
```

```python
print("State Unique Values:", data['State'].unique())
```

### GPA 

```python
print("GPA Unique Values:", data['GPA'].unique())
```

```python
data["GPA"] = data["GPA"].where(data["GPA"] <= 4, np.nan)
print("GPA Unique Values:", data['GPA'].unique())
data["GPA"] = data["GPA"].astype(float)
```

```python
data = data.dropna()
data
```

### WorkExp

```python
print("Work Experience Unique Values:", data['WorkExp'].unique())
data["WorkExp"] = data["WorkExp"].where(data["WorkExp"] <= 10, np.nan)
data = data.dropna()
print("Work Experience Unique Values:", data['WorkExp'].unique())
data["WorkExp"] = data["WorkExp"].astype(float)
```

### TestScore

```python
print("Test Score Unique Values:", data['TestScore'].unique())
data["TestScore"] = data["TestScore"].astype(float)
data
```

### Writing Score

Taking out the writing score of 1 because by the IQR rule, it is an outlier. 

```python
print("Writing Score Unique Values:", data["WritingScore"].unique())
data["WritingScore"] = data["WritingScore"].where(data["WritingScore"] >= 2, np.nan)
data = data.dropna()
print("Writing Score Unique Values:", data["WritingScore"].unique())
data["WritingScore"] = data["WritingScore"].astype(float)
```

## Gender 

```python
print("Gender Unique Values:", data['Gender'].unique())
data = data[data["Gender"] != -1]
```

```python
print("Gender Unique Values:", data['Gender'].unique())
data["Gender"] = data["Gender"].astype(int)
```

### Volunteer Level

```python
data["VolunteerLevel"].unique()
```

```python
from pandas.api.types import CategoricalDtype

categories = [0, 1, 2, 3, 4, 5]
data["VolunteerLevel"] = data["VolunteerLevel"].astype(CategoricalDtype(categories, ordered=True))
data
```
